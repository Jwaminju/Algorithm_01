```python
fix, ax, = plt.subplots(2,3)
ax[0,0].plot([1,2,3],[1,2,3]) # index로 접근가능하다 subplot(221)처럼 안해도 됨
ax[0,1].plot([1,2,3],[1,2,3])
ax[0,1].fill_between([1,2,3][0:2],[1,2,3][0:2], color='y', alpha=0.2) # 투명도
ax[0,2].plot([1,2,3],[1,2,3])
ax[0,2].fill_betweenx([1,2,3][0:2],[1,2,3][0:2], color='y', alpha=0.2) # 투명도
plt.savefig('/image.png')
plt.show()
```

DataFrame을 이용한 linechar 그리기

```python
df = pd.DataFrame(np.random.randn(50,3), columns = ['A', 'B', 'C'], index = list(range(1, 51)))
```

```python
df
```

```python
plt.plot(df)
plt.legend(df, title = 'group', loc = 'best') # location
```

```python
fig, ax1 = plt.subplots()

line1 = ax1.plot(df.A, label = 'A')
line2 = ax1.plot(df.B, label = 'B')
ax1.set_ylabel('A,B')
ax1.set_ylim(0,5)

ax2 = ax1.twinx()
line3 = ax2.plot(df.C, label = 'C', color='m')
ax2.set_ylabel('C')
ax2.set_ylim(0,10)

lines = line1 + line2 + line3

labels = [i.get_label() for i in lines]
plt.legend(lines, labels, loc =  'upper left')
plt.show()
```

```python
plt.hist(df.A, bins = 7)
plt.show()
```

```python
bins1 = range(-3, 3, 1) # range는 실수가 아니라 정수간격
plt.hist(df.A, bins = bins1)
plt.xticks(bins1)
plt.show()
```

```python
bins1 = range(-3, 3, 1) # range는 실수가 아니라 정수간격
plt.hist(df.A, bins = bins1, label = 7)
plt.hist(df.A, bins = 7, label = 10, color = 'r', alpha = 0.5)
plt.xticks(bins1)
plt.legend(title = 'bins')
plt.show()
```

```python
plt.hist((df.A, df.B), label = [10,7], color = ['r', 'b'], alpha = 0.5, histtype='bar')
plt.xticks(bins1)
plt.legend(title = 'bins')
plt.show()
```

```python
plt.hist((df.A, df.B), label = [10,7], color = ['r', 'b'], alpha = 0.5, histtype='barstacked')
plt.legend()
plt.show()
```

```python
plt.hist((df.A, df.B), label = [10,7], color = ['r', 'b'], alpha = 0.5, histtype='step')
plt.legend()
plt.show()
```

```python
plt.hist((df.A, df.B), label = [10,7], color = ['r', 'b'], alpha = 0.5, histtype='stepfilled')
plt.legend()
plt.show()
```

```python
df_new = df.loc[(df.A>=0) & (df.B >=0)]
plt.hist((df.A, df.B), label = [10,7], color = ['r', 'b'], alpha = 0.5, histtype='stepfilled')
plt.legend()
plt.show()
```
