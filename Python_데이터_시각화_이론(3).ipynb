```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

상자 그래프:

이상값(IQR 이용해서 계산), 최대값, 최소값, 중앙값(not 평균), 제 1사분위 수 Q1(데이터 25%), 제 3사분위 수 Q3(데이터 75% 지점)

최소값: Q1 - 1.5*IQR 보다 큰 데이터 중 가장 작은 값

최대값: Q3 + 1.5*IQR 보다 작은 데이터 중 가장 큰 값

이상값:최소값/최대값보다 작은/큰 데이터


```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips.total_bill, tips.tip)
plt.show()
```

```python
plt.scatter(tips.total_bill, tips.tip, s=5, c='r')
plt.show()
```

seaborn으로 산점도 그리기(다양한 모양이 많다)

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day') # 요일별로 범주 구분 가능 -> 요일은 범주형 자료 <-> 실수형은 연속적인 색상으로 표현됨
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue="size", size="size", sizes=(5,200)) # 사이즈별로 점 사이즈
plt.show()
```

```python
plt.boxplot(tips.tip, sym='b')# 이상치의 색도 바꾸기 가능
plt.show()
```

```python
plt.boxplot([tips.total_bill, tips.tip], sym='b*')# 이상치의 색도 바꾸기 가능
plt.xticks([1,2], ['total_bill', 'tip'])
plt.show()
```

```python
sns.boxplot(x='day', y='tip', data=tips)
plt.show()
```

swarmplot: 데이터가 어디에 많이 분포되어 있는지 알려준다

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex', size=3)
plt.show() # 경고문이 나오는 것은 점 사이즈 때문에 다 담을 수 없어서 꺾이기 때문이다 -> size 조정
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex', size=3)
plt.show() # 경고문이 나오는 것은 점 사이즈 때문에 다 담을 수 없어서 꺾이기 때문이다 -> size 조정
```
